```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
cases_url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
Covid_cases=pd.read_csv(cases_url)
```

```python
recovered_url='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'
Covid_recover=pd.read_csv(recovered_url)
```

```python
Covid_cases['Country/Region'].replace('Mainland China', 'China', inplace=True)
Covid_deaths['Country/Region'].replace('Mainland China', 'China', inplace=True)
Covid_recover['Country/Region'].replace('Mainland China', 'China', inplace=True)
```

```python
Covid_cases[['Province/State']] = Covid_cases[['Province/State']].fillna('')
Covid_cases.fillna(0, inplace=True)

Covid_deaths[['Province/State']] = Covid_deaths[['Province/State']].fillna('')
Covid_deaths.fillna(0, inplace=True)

Covid_recover[['Province/State']] = Covid_recover[['Province/State']].fillna('')
Covid_recover.fillna(0, inplace=True)
```

```python
Covid_recover.head(10)
```

```python
Cov_cases_sum = Covid_cases.iloc[:, 4:].sum().max()
```

```python
Cov_cases_sum
```

```python
Cov_deaths_sum = Covid_deaths.iloc[:, 4:].sum().max()
```

```python
Cov_deaths_sum
```

```python
Cov_recover_sum = Covid_recover.iloc[:, 4:].sum().max()
```

```python
Cov_recover_sum
```

```python
world_cases = pd.DataFrame({
    'confirmed': [Cov_cases_sum],
    'deaths': [Cov_deaths_sum],
    'recovered': [Cov_recover_sum],
    'active': [Cov_cases_sum - Cov_deaths_sum - Cov_recover_sum]
})
```

```python
world_cases
```

```python
world_cases_confirmed = world_cases.melt(value_vars=['active', 'deaths', 'recovered'],
                              var_name="status",
                              value_name="count")

world_cases_confirmed['upper'] = 'confirmed'

world_cases_confirmed
```

```python
covid_worldwide_confirmed = Covid_cases.iloc[:, 4:].sum(axis=0)
```

```python
covid_worldwide_confirmed.head()
```

```python
covid_worldwide_deaths = Covid_deaths.iloc[:, 4:].sum(axis=0)
```

```python
covid_worldwide_deaths.head()
```

```python
covid_worldwide_recovered = Covid_recover.iloc[:, 4:].sum(axis=0)
```

```python
covid_worldwide_recovered.head()
```

```python
covid_worldwide_active = covid_worldwide_confirmed - covid_worldwide_deaths - covid_worldwide_recovered
```

```python
covid_worldwide_active.head()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.lineplot(x=covid_worldwide_confirmed.index, y=covid_worldwide_confirmed, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_deaths.index, y=covid_worldwide_deaths, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_recovered.index, y=covid_worldwide_recovered, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_active.index, y=covid_worldwide_active, sort=False, linewidth=2)

ax.lines[0].set_linestyle("--")

plt.suptitle("COVID-19 worldwide cases over the time evolution", fontsize=16, fontweight='bold', color='white')

plt.xticks(rotation=45)
plt.ylabel('Number of cases')

ax.legend(['Confirmed', 'Deaths', 'Recovered', 'Active'])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
ax.set(yscale="log")
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))

sns.lineplot(x=covid_worldwide_confirmed.index, y=covid_worldwide_confirmed, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_deaths.index, y=covid_worldwide_deaths, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_recovered.index, y=covid_worldwide_recovered, sort=False, linewidth=2)
sns.lineplot(x=covid_worldwide_active.index, y=covid_worldwide_active, sort=False, linewidth=2)

ax.lines[0].set_linestyle("--")

plt.suptitle("COVID-19 worldwide cases over the time", fontsize=16, fontweight='bold', color='white')
plt.title("(logarithmic scale)", color='white')

plt.xticks(rotation=45)
plt.ylabel('Number of cases')

ax.legend(['Confirmed', 'Deaths', 'Recovered', 'Active'])

plt.show()
```
